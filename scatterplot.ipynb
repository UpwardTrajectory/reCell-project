```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


uniform = np.random.uniform(1000, size=3000)


def create_graphs(arr):
    plt.figure(figsize=(15, 30))

    plt.subplot(3, 1, 3)
    sns.histplot(arr)

    plt.subplot(3, 1, 2)
    sns.boxplot(x=arr)

    plt.subplot(3, 1, 1)
    xs = np.linspace(0, max(arr), len(arr))
    sns.scatterplot(x=xs, y=sorted(arr))

    plt.tight_layout()
    return plt


create_graphs(uniform).show()
```

```python
exponential = np.random.exponential(1000.0, size=3000)
create_graphs(exponential)
```

```python
logged = np.log(exponential)

create_graphs(logged)
```
